# While

La primera estructura de iteración que se verá aquí es la sentencia *while*, que significa *mientras*. Con la estructura mientras, le decimos que repita una acción, una y otra vez, hasta que se cumpla la condición.

En Python, while tiene el significado  *mientras se cumpla esta condición, repite estas acciones*. Las sentencias que denotan estas repeticiones se denominan 

```python
contador=0
while contador<=10:
    print(contador)
```

```python
contador=0 #condicion inicial
while contador<=10:
    contador=contador+1
    print(contador)
```

Hacer un programa que calcule la sumatoria de Gauss:

$$
\sum_{i=1}^{100}i = 1+2+3+ \cdots +100
$$


```python
sumatoria=0
i=0  #contador 
while i<=100:
    sumatoria += i
    i+=1
print("La suma es",sumatoria)
```

```python
sumatoria=0
i=0
while i<=100:
    sumatoria += i**2
    i+=2 #avanzamos i de dos en dos
print("La suma es",sumatoria)
```

Serie de Fibonacci, que regresa la secuencia de números 


0, 1, 1, 2, 3, 5, 8, 13, 21 ...

```python
a, b = 0, 1
while b < 30:
    print(b)
    a,b = b, a+b
```

Otra variación muy interesante de la estructura while es incluir la sentencia break, la cual nos permite romper abruptamente el ciclo. Un ejemplo muy requerido en cómputo científico es crear un menú, por lo cual usaremos while-break.

```python
print('Menu')
while 1:  #ciclo infinito
    opcion=int(input('Escribe la opción que gustes, para finalizar usa 0 \t'))
    if opcion==1:
        print('Primera opción')
    elif opcion==2:
        print('Segunda opción')
    elif opcion==0:
        print('Adios')
        break  #rompemos el bucle o ciclo infinito
    else:
        print('da una opción valida')
```

Esta estructura es un poco truculenta. Usamos un bucle infinito, que el usuario romperá hasta que entre a la opción que incluye break.
